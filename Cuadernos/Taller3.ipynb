### Escala de perspectiva a futuro (FTP)

1. [La escala de perspectiva de tiempo futuro](https://academic.oup.com/gerontologist/article/54/6/1075/2957455?login=true#57814318):  es un cuestionario de 12 ítems evaluado en una escala Likert de cinco puntos (1 = totalmente en desacuerdo, 5 = totalmente de acuerdo). Mide tres perspectivas del futuro: Future as Open (4 elementos), Future as Limited (4 elementos) y Future as Ambiguous (4 elementos). El alfa de Cronbach para las tres dimensiones en el presente estudio fue de 0,85, 0,74 y 0,80, respectivamente. Ejemplos de Escala de Perspectiva de Tiempo Futuro son: "Espero el futuro con esperanza y entusiasmo" (Futuro como Abierto), "Cada vez más siento que el tiempo está en mi contra" (Futuro como Limitado) y "No me enfoco en el futuro". porque es tan incierto para mí de todos modos” (Futuro como ambiguo).

2. Las preguntas que conforman esta escala son las siguientes:

- (**FTP1**) I look forward to the future with hope and enthusiasm
- (**FTP2**) When I think about the future, I expect good things to happen	
-  (**FTP3**) I have little hope for the future
-  (**FTP4**) Each new day and season presents me with interesting opportunities
-  (**FTP5**) I have the sense that time is running out	
- (**FTP6**)  I am beginning to experience that time is limited	
-  (**FTP7**) Increasingly I feel like time is against me
-  (**FTP8**) I know that I do not have all the time in the world
- (**FTP9**)  My future is uncertain
-  (**FTP10**) You cannot really plan for the future because things change so much
- (**FTP11**)  My future seems very vague and uncertain to me
- (**FTP12**) I do not focus on the future because it is so uncertain to me anyway
